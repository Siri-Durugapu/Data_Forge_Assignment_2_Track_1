```python
import time
import mysql.connector

db = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Swamini@64",
    database="scholarease"
)

cursor = db.cursor()

# BEFORE INDEX
start = time.time()
cursor.execute("SELECT * FROM member WHERE Name='Rahul'")
cursor.fetchall()
end = time.time()

before_time = end - start

# AFTER INDEX (run after creating index)
start = time.time()
cursor.execute("SELECT * FROM member WHERE Name='Rahul'")
cursor.fetchall()
end = time.time()

after_time = end - start

print("Before:", before_time)
print("After:", after_time)

import matplotlib.pyplot as plt

labels = ['Before Index', 'After Index']
times = [before_time, after_time]

plt.bar(labels, times)
plt.xlabel("Query Type")
plt.ylabel("Execution Time (seconds)")
plt.title("Indexing Performance Comparison")

plt.show()
```
